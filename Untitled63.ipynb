```python
import numpy as np
import matplotlib.pyplot as plt

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

# Sample Data
x1 = 3
x2 = 2
y_actual = 17

# Initialize random weights
w1 = np.random.uniform(1, 10)
w2 = np.random.uniform(1, 10)

# Hyperparameters
learning_rate = 0.01
epochs = 20

# Lists to store error and predictions across epochs
errors = []
ypreds = []

# Training
for epoch in range(epochs):
    # Feed Forward
    y_pred = x1 * w1 + x2 * w2
    ypreds.append(y_pred)

    # Error Calculation
    error = (y_actual - y_pred) ** 2
    errors.append(error)

    # Back Propagation
    # Compute gradients
    grad_w1 = -2 * x1 * (y_actual - y_pred)
    grad_w2 = -2 * x2 * (y_actual - y_pred)

    # Update weights
    w1 -= learning_rate * grad_w1
    w2 -= learning_rate * grad_w2

# Visualization
# Plot progression of error across epochs
plt.plot(range(epochs), errors)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.title('Progression of Error Across Epochs')
plt.show()

# Plot progress of y_pred across epochs
plt.plot(range(epochs), ypreds)
plt.xlabel('Epochs')
plt.ylabel('ypred')
plt.title('Progress of y_pred Across Epochs')
plt.show()
```
